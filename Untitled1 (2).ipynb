```python
import gradio as gr
from PIL import Image, ImageEnhance, ImageFilter
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler, AutoencoderKL
import os
import gc
# CORRECTED IMPORT FOR GEMINI
import google.generativeai as genai
import random

# --- GEMINI API CONFIGURATION ---
# NOTE: Apni asal Gemini API key yahan "YOUR_GEMINI_API_KEY" ki jagah daalein.
# Aap isay secrets mein ya environment variable ke tor par bhi rakh sakte hain.
GEMINI_API_KEY = "AIzaSyBG41qdwjb2XcX-lPy67Nu1uUBofkUghF8"

try:
    genai.configure(api_key=GEMINI_API_KEY)
    GEMINI_AVAILABLE = True
    print("Gemini API configured successfully.")
except Exception as e:
    GEMINI_AVAILABLE = False
    print(f"WARNING: Could not configure Gemini API. AI Prompt Helper will be disabled. Error: {e}")
# --------------------------------

# Global variables to hold the pipeline and track the current LoRA
global_pipe = None
current_lora = None

def load_model_sdxl_lite(model_path_sdxl, model_path_lora=None):
    print("Loading model...")
    vae = AutoencoderKL.from_pretrained(
        "madebyollin/sdxl-vae-fp16-fix",
        torch_dtype=torch.float16
    )
    pipe = DiffusionPipeline.from_pretrained(
        model_path_sdxl,
        vae=vae,
        torch_dtype=torch.float16,
        variant="fp16",
        use_safetensors=True
    )

    if model_path_lora and not os.path.exists(model_path_lora):
        import wget
        print("Downloading LoRA weights...")
        if "Logo" in model_path_lora:
            lora_url = "https://huggingface.co/artificialguybr/LogoRedmond-LogoLoraForSDXL-V2/resolve/main/LogoRedmondV2-Logo-LogoRedmAF.safetensors"
        else:
            lora_url = "https://huggingface.co/artificialguybr/IconsRedmond-IconsLoraForSDXL/resolve/main/IconsRedmondV2-Icons.safetensors"
        wget.download(lora_url, model_path_lora)

    if model_path_lora and os.path.exists(model_path_lora):
        print(f"Loading LoRA from: {model_path_lora}")
        pipe.load_lora_weights(model_path_lora)
        pipe.fuse_lora()

    device = "cuda" if torch.cuda.is_available() else "cpu"
    pipe.to(device)
    pipe.enable_vae_slicing()
    pipe.enable_vae_tiling()

    try:
        import xformers
        pipe.enable_xformers_memory_efficient_attention()
    except ImportError:
        print("xformers not found.")

    if device == "cpu" or torch.cuda.get_device_properties(0).total_memory < 8e9:
        pipe.enable_model_cpu_offload()

    pipe.scheduler = DPMSolverMultistepScheduler.from_config(
        pipe.scheduler.config,
        use_karras_sigmas=True,
        algorithm_type="dpmsolver++"
    )
    print("Model loaded.")
    return pipe

def generate_image_colab(pipe, prompt, negative_prompt=None, steps=50, cfg=7.5, width=1024, height=1024):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    if pipe.device.type != device:
        pipe.to(device)
    with torch.no_grad():
        output = pipe(prompt=prompt, negative_prompt=negative_prompt, num_inference_steps=steps, guidance_scale=cfg, width=width, height=height, eta=0.0)
        return output.images[0]

def enhance_image(image, quality_mode="high"):
    if image.mode != 'RGB':
        image = image.convert('RGB')
    if quality_mode == "high":
        image = ImageEnhance.Sharpness(image).enhance(1.1)
        image = ImageEnhance.Contrast(image).enhance(1.0)
        image = ImageEnhance.Color(image).enhance(1.0)
        width, height = image.size
        new_width, new_height = int(width * 1.5), int(height * 1.5)
        image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)
        image = image.filter(ImageFilter.UnsharpMask(radius=1.0, percent=150, threshold=3))
    elif quality_mode == "ultra":
        image = ImageEnhance.Sharpness(image).enhance(1.1)
        image = ImageEnhance.Contrast(image).enhance(1.05)
        width, height = image.size
        new_width, new_height = int(width * 2), int(height * 2)
        image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)
    return image

def generate_logos_gradio(prompt, negative_prompt, num_images,
                          design_type, logo_shape, style_type, color_scheme,
                          quality_mode, resolution):
    global global_pipe, current_lora
    model_path = "stabilityai/stable-diffusion-xl-base-1.0"
    if design_type == "Logo":
        lora_path = "/content/LogoRedmondV2-Logo-LogoRedmAF.safetensors"
        requested_lora = "Logo"
    else:
        lora_path = "/content/IconsRedmondV2-Icons.safetensors"
        requested_lora = "Icon"

    if global_pipe is None or current_lora != requested_lora:
        if global_pipe is not None:
            print(f"Switching from '{current_lora}' to '{requested_lora}' LoRA. Reloading pipeline...")
            del global_pipe
            gc.collect()
            torch.cuda.empty_cache()
        print(f"Loading model with '{requested_lora}' LoRA...")
        try:
            global_pipe = load_model_sdxl_lite(model_path, lora_path)
            current_lora = requested_lora
        except Exception as e:
            print(f"Error loading model: {e}")
            return [Image.new("RGB", (512, 512), "gray")]

    resolution_settings = {"Standard (1024x1024)": (1024, 1024, 50, 7.5), "High (1152x1152)": (1152, 1152, 60, 8.0), "Ultra (1280x1280)": (1280, 1280, 70, 8.5)}
    width, height, steps, cfg = resolution_settings[resolution]
    shape_prompts = {"Circle": "circular logo, round emblem, perfect circle", "Square": "square logo, rectangular frame, geometric design", "Shield": "shield logo, badge design, protective emblem", "Hexagon": "hexagonal logo, six-sided geometric shape", "Diamond": "diamond-shaped logo, rhombus design", "Custom": ""}
    style_prompts = {"Modern": "modern design, clean lines, contemporary style, minimalist", "Classic": "classic design, traditional style, elegant, timeless", "Tech": "futuristic design, high-tech style, digital, sleek", "Vintage": "vintage design, retro style, classic aesthetic", "Creative": "creative design, artistic style, unique, expressive"}
    color_prompts = {"Professional": "professional color scheme, blue and gray tones", "Vibrant": "vibrant colors, bright and energetic palette", "Monochrome": "black and white, monochrome, grayscale", "Luxury": "luxury colors, gold and black, premium palette", "Natural": "natural colors, green and earth tones", "Custom": ""}

    images = []
    for i in range(num_images):
        if design_type == "Logo":
            current_prompt = f"professional logo design of {prompt}, vector style, clean design, high resolution, 4K, ultra HD, crisp edges, studio lighting, brand identity"
        else:
            current_prompt = f"app icon design of {prompt}, mobile icon style, clean interface design, high resolution, 4K, ultra HD, crisp edges, user interface icon"
        if logo_shape != "Custom":
            current_prompt += f", {shape_prompts[logo_shape].replace('logo', 'icon') if design_type == 'Icon' else shape_prompts[logo_shape]}"
        current_prompt += f", {style_prompts[style_type]}"
        if color_scheme != "Custom":
            current_prompt += f", {color_prompts[color_scheme]}"
        if i > 0:
            current_prompt += f", {random.choice(['alternative design', 'different approach', 'creative variation'])}"

        enhanced_negative = f"{negative_prompt}, blurry, low resolution, pixelated, poorly drawn, distorted, watermark, text, signature, bad anatomy, deformed, broken shapes, photographic, photo-realistic, sketchy, abstract clutter, messy lines, oversaturated, overly detailed, human face, hands, body parts, background clutter, noise, artifacts, unprofessional, realism, low quality, gradient banding, chromatic aberration, depth-of-field blur, out of focus"
        if design_type == "Icon":
            enhanced_negative += ", too detailed, cluttered"

        try:
            print(f"Generating {design_type.lower()} {i+1}/{num_images}...")
            img = generate_image_colab(global_pipe, current_prompt, enhanced_negative, steps=steps, cfg=cfg, width=width, height=height)
            enhanced = enhance_image(img, quality_mode)
            images.append(enhanced)
        except RuntimeError as e:
            if 'CUDA out of memory' in str(e):
                print("CUDA out of memory, trying with reduced resolution...")
                torch.cuda.empty_cache()
                gc.collect()
                try:
                    img = generate_image_colab(global_pipe, current_prompt, enhanced_negative, steps=40, cfg=7.0, width=768, height=768)
                    enhanced = enhance_image(img, "high")
                    images.append(enhanced)
                except Exception as fallback_error:
                    print(f"Fallback failed: {fallback_error}")
                    images.append(Image.new("RGB", (512, 512), "red"))
            else:
                print(f"Generation error: {e}")
                images.append(Image.new("RGB", (512, 512), "red"))
        torch.cuda.empty_cache()
        gc.collect()
    return images

# --- CORRECTED GEMINI FUNCTION ---
def generate_prompt_with_gemini(scenario_text, design_type):
    if not GEMINI_AVAILABLE:
        return "Gemini API key not configured or invalid. Please check the `GEMINI_API_KEY` in the code."
    if not scenario_text or not scenario_text.strip():
        return ""

    if design_type == "Logo":
        system_instruction = "You are a prompt engineer for designing professional logos for Stable Diffusion XL. Convert the user's idea into a visually descriptive logo prompt. Focus on symbolism, modern design styles (vector, flat, minimalist), and brand personality. Avoid photographic or cluttered terms. Example: 'A minimalist logo for a tech startup, featuring a stylized brain icon with interconnected nodes, symbolizing intelligence and connectivity, clean vector art.'"
    else:  # Icon
        system_instruction = "You are a prompt engineer for designing app icons for Stable Diffusion XL. Convert the user's app idea into a simple, recognizable icon prompt. Focus on the core function, using clean symbols suitable for mobile interfaces. Avoid complex details. Example: 'An app icon for a weather app, showing a stylized sun partially covered by a fluffy white cloud, simple flat design, vibrant colors.'"

    try:
        model = genai.GenerativeModel(
            model_name="gemini-1.5-flash-latest",
            system_instruction=system_instruction
        )
        response = model.generate_content(scenario_text)
        return response.text.strip()
    except Exception as e:
        print(f"Error generating prompt with Gemini: {e}")
        return f"Error from Gemini API: {str(e)}"
# ---------------------------------

def update_examples_and_prompt(design_type):
    if design_type == "Logo":
        prompt_placeholder = "e.g., modern tech company with clean geometric design"
    else:
        prompt_placeholder = "e.g., weather app with sun and cloud symbols"
    return gr.update(placeholder=prompt_placeholder)

# Gradio UI (No changes needed here)
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 🎨 Professional Logo & Icon Generator")
    gr.Markdown("Generate high-quality, professional logos and app icons with simple customization")
    with gr.Column():
        with gr.Group():
            gr.Markdown("### 🎯 Choose Design Type")
            design_type = gr.Radio(choices=["Logo", "Icon"], label="What do you want to create?", value="Logo", info="Logo: Brand identity | Icon: App icons, UI elements")
        with gr.Group():
            gr.Markdown("### 🧠 AI Prompt Helper")
            scenario = gr.Textbox(label="Describe Your Business/App", placeholder="e.g., A sustainable energy company that focuses on solar power", lines=2)
            generate_prompt_btn = gr.Button("✨ Generate Prompt", variant="secondary")
        with gr.Group():
            gr.Markdown("### ✏️ Design Description")
            prompt = gr.Textbox(label="What should your design represent?", value="modern tech company with clean geometric design", lines=2)
            negative_prompt = gr.Textbox(label="What to avoid in the design", value="blurry, low resolution, pixelated, poorly drawn, distorted, watermark, text, signature, bad anatomy, unrealistic, ugly, broken shapes, deformed, photographic, 3D render, sketchy, amateur, unprofessional, low quality, artifacts, noise", lines=2)
        with gr.Group():
            gr.Markdown("### 🎨 Design Style")
            with gr.Row():
                logo_shape = gr.Dropdown(choices=["Custom", "Circle", "Square", "Shield", "Hexagon", "Diamond"], label="Shape", value="Custom", info="Choose a specific shape or let AI decide")
                style_type = gr.Dropdown(choices=["Modern", "Classic", "Tech", "Vintage", "Creative"], label="Design Style", value="Modern", info="Overall design approach")
                color_scheme = gr.Dropdown(choices=["Custom", "Professional", "Vibrant", "Monochrome", "Luxury", "Natural"], label="Color Theme", value="Custom", info="Color palette for your design")
        with gr.Group():
            gr.Markdown("### ⚙️ Generation Settings")
            with gr.Row():
                num_images = gr.Slider(1, 4, step=1, label="Number of Designs", value=2)
                quality_mode = gr.Dropdown(choices=["high", "ultra"], label="Quality", value="high", info="Ultra = best quality but slower")
                resolution = gr.Dropdown(choices=["Standard (1024x1024)", "High (1152x1152)", "Ultra (1280x1280)"], label="Resolution", value="Standard (1024x1024)", info="Higher = better quality but more VRAM")
        generate = gr.Button("🚀 Generate Designs", variant="primary", size="lg")
    with gr.Group():
        gr.Markdown("### 📸 Your Generated Designs")
        gallery = gr.Gallery(label="Professional Design Results", columns=2, height="auto", show_download_button=True, show_share_button=False, preview=True)

    # Event handlers
    generate_prompt_btn.click(fn=generate_prompt_with_gemini, inputs=[scenario, design_type], outputs=prompt)
    design_type.change(fn=update_examples_and_prompt, inputs=design_type, outputs=prompt)
    generate.click(fn=generate_logos_gradio, inputs=[prompt, negative_prompt, num_images, design_type, logo_shape, style_type, color_scheme, quality_mode, resolution], outputs=gallery)

if __name__ == "__main__":
    demo.launch(share=True, debug=True)
```
